```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def preorder(root):
    if root==None:
        return
    print(root.data,end=' ')
    preorder(root.left)
    preorder(root.right)
def inorder(root):
    if root==None:
        return
    inorder(root.left)
    print(root.data,end=' ')
    inorder(root.right)
def postorder(root):
    if root==None:
        return
    postorder(root.left)
    postorder(root.right)
    print(root.data,end=' ')
def BFS(root):
    if root is None:
        return
    queue=[]
    queue.append(root)
    while(len(queue) > 0):
        print(queue[0].data, end=" ")
        node = queue.pop(0)
        if node.left is not None:
            queue.append(node.left)
        if node.right is not None:
            queue.append(node.right)
def levelOrder(root):
    if (root == None):
        return
    q = queue()
    q.append(root)
    q.append(None)
    while (len(q) > 1):
        curr = q.popleft()
        if (curr == None):
            q.append(None)
            print()
        else:
            if (curr.left):
                q.append(curr.left)
            if (curr.right):
                q.append(curr.right)
            print(curr.data, end=" ")
def Bfs_2(root):#sir ka method
    q=[root]
    q.append(None)
    while len(q)>0:
        curr=q.pop(0)
        if curr==None:
            if len(q)==0:
                break
            else:
                print()
                q.append(None)
        else:
            print(curr.data,end=' ')
            if curr.left!=None:
                q.append(curr.left)
            if curr.right!=None:
                q.append(curr.right)     
def height(root):
    if root==None:
        return 0
    l=height(root.left)
    r=height(root.right)
    return max(l,r)+1
def printLeaf(root):
    if root is None:
        return
    if root.left is None and root.right is None:
        print(root.data,end=" ")
    if root.left:
        printLeaf(root.left)
    if root.right:
        printLeaf(root.right)
if __name__=="__main__":
    root=Node(1)
    root.left=Node(2)
    root.right=Node(3)
    root.left.left=Node(4)
    root.left.right=Node(5)
    root.right.left=Node(6)
    root.right.right=Node(7)
    print("preorder:")
    preorder(root)
    print()
    print("inorder:")
    inorder(root)
    print()
    print("preorder:")
    postorder(root)
    print("\nlevel order")
    BFS(root)
    print("\nlevel order")
    Bfs_2(root)
    print()
    print("Height=",height(root))
    print("leaf nodes are")
    printLeaf(root)
    
```

```python
class Node:
    def __init__(self, key):
        self.data = key
        self.left = None
        self.right = None

def top_view(root):
    if root is None:
        return

    # This will store the horizontal distance of nodes
    hd_node = {}

    # Queue for level order traversal
    queue = []

    # Horizontal distance of root is 0
    queue.append((root, 0))

    while len(queue):
        node, hd = queue.pop(0)

        # If this is the first node at its horizontal distance, add it to the result
        if hd not in hd_node:
            hd_node[hd] = node.data

        # Enqueue left child with horizontal distance hd-1
        if node.left:
            queue.append((node.left, hd - 1))

        # Enqueue right child with horizontal distance hd+1
        if node.right:
            queue.append((node.right, hd + 1))

    # Extracting the entries of the hd_node dictionary and printing them in sorted order of keys
    for hd in sorted(hd_node):
        print(hd_node[hd], end=' ')

# Example Usage
if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.right = Node(4)
    root.left.right.right = Node(5)
    root.left.right.right.right = Node(6)
    print("Top view of the binary tree is: ")
    top_view(root)
```
